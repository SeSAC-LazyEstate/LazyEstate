```python
# 기본 폴더 경로 설정
import os
import csv
import chardet

# base_folder = os.getcwd()+'/서울시'  # 서울시의 여러 구 별로 폴더가 생성되어 있는 기본 폴더 경로
base_folder = '/Users/sopung/Desktop/output/서울시'  # 서울시의 여러 구 별로 폴더가 생성되어 있는 기본 폴더 경로

# 모든 구 폴더를 순회하면서 동 별 CSV 파일을 구 별로 합치는 함수
def combine_csv_files_by_district(base_folder):
    district_folders = [folder for folder in os.listdir(base_folder) if os.path.isdir(os.path.join(base_folder, folder))]
    
    if not district_folders:
        print("구 폴더가 없습니다.")
        return
    
    for district in district_folders:
        district_output_file = f"{district}.csv"  # 각 구 이름으로 저장할 CSV 파일 이름
        district_output_path = os.path.join(base_folder, district_output_file)
        
        with open(district_output_path, 'w', newline='', encoding='utf-8') as district_csv:  # 파일의 실제 인코딩을 설정
            csv_writer = csv.writer(district_csv)
            
            # 필요한 칼럼명들을 추가
            column_names = [
                "순번", "아파트명", "면적", "법정동주소", "b_code", "지번주소", "도로명주소", "latitude", "longitude",
                "세대수", "임대세대수", "최고층", "최저층", "용적률", "건폐율", "주차대수",
                "건설사", "난방", "공급면적", "전용면적", "전용율", "방수", "욕실수",
                "해당면적_세대수", "현관구조", "재산세", "재산세합계", "지방교육세", "재산세_도시지역분",
                "종합부동산세", "결정세액", "농어촌특별세", "가격", "겨울관리비", "여름관리비",
                "매매호가", "전세호가", "월세호가", "실거래가", "초등학교_학군정보",
                "초등학교_설립정보", "초등학교_남학생수", "초등학교_여학생수"
            ]
            csv_writer.writerow(column_names)
            
            # 구 폴더 내의 동 별 CSV 파일들을 순회하며 데이터를 합침
            district_folder_path = os.path.join(base_folder, district)
            csv_files = [file for file in os.listdir(district_folder_path) if file.endswith('.csv')]
            
            for csv_file in csv_files:
                csv_path = os.path.join(district_folder_path, csv_file)
                
                # 파일의 실제 인코딩을 자동으로 판별하여 설정
                with open(csv_path, 'rb') as csv_content:
                    detected_encoding = chardet.detect(csv_content.read())['encoding']
                
                with open(csv_path, 'r', encoding=detected_encoding) as csv_content:
                    csv_reader = csv.reader(csv_content)
                    next(csv_reader)  # 헤더는 제외
                    
                    for row in csv_reader:
                        csv_writer.writerow(row)
                
                print(f"{csv_file} 파일이 {district_output_file}에 합쳐짐")
    
        print(f"{district_output_file} 파일이 생성되었습니다.")

# CSV 파일 합치기 함수 실행
combine_csv_files_by_district(base_folder)
```
